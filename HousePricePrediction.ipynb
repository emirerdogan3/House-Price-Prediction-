### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
df2 = dfCleaned.drop(["street","city","statezip","country","date","yr_renovated"],axis=1)
```

```python
df2.info()
```

```python
y = df2["price"].values
y
```

```python
x = df2[["bedrooms","bathrooms","sqft_living","sqft_lot","floors","waterfront","view","condition","sqft_above","sqft_basement","yr_built"]].values
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
# x_train = scaler.fit_transform(x_train)
x_train
```

```python
x_test = scaler.transform(x_test)
# x_test = scaler.fit_transform(x_test)
x_test
```

### Modeli Oluşturma

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(units=30, activation="relu")) # 30 sütun olduğu için 30 ile başladık
model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))

model.add(Dense(units=1)) # output layeri

model.compile(optimizer="adam", loss="mse")
# model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", patience=25, mode="min", verbose=1)
```

```python
model.fit(x=x_train, y=y_train, epochs=300, verbose=1, validation_data=(x_test, y_test),callbacks=[earlyStopping])
```

```python
pd.DataFrame(model.history.history)
```

```python
modelKaybi = pd.DataFrame(model.history.history)
plt.plot(modelKaybi)
```

```python
from tensorflow.keras.models import load_model
```

```python
model.save("hausePricePredictionModel.h5")
```
